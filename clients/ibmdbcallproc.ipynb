```python
import ibm_db
import getpass

user = "userid"
pwd  = getpass.getpass()

targetdb2 = "DATABASE=<DB2_LOCATION>;" + \
                    "HOSTNAME=<db2-ip>;" + \
                    "PORT=<db2-port>;" + \
                    "PROTOCOL=TCPIP;" + \
                    "UID=" + user + ";" + \
                    "PWD=" + pwd + ";"
```

```python
db = 'DBIDUG'
ts = 'TSIDUG01'
rc = 0
lfdnr = 0
```

```python
conn = ibm_db.connect(targetdb2, "", "")
erg = ibm_db.callproc(conn, 'IDUG.QUIESCETS', (db, ts, rc, lfdnr))
result = erg[0]                     # Result in 1st
row = ibm_db.fetch_both(result)
while(row):
    print(row)         
    row = ibm_db.fetch_both(result)


# Python with ibm_db:
# https://www.ibm.com/docs/en/ias?topic=framework-application-development-db

# Stackoverflow Reading ibm_db Result Set: (fetch_tuple, fetch_assoc, fetch_both, fetch_row)
# https://stackoverflow.com/questions/49620086/ibm-db-execute-how-to-get-the-result-set

# Close Connection
rc = ibm_db.close(conn)
print(rc)
```
